```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/datasets/gdp/master/data/gdp.csv"
df = pd.read_csv(url)

# Filter for Israel
israel_gdp = df[df['Country Name'] == 'Israel']

#print(israel_gdp.head())
display(israel_gdp.head())
```

```python

display(israel_gdp.isnull().sum())

display(israel_gdp.info())
```

```python
# Save the israel_gdp DataFrame to a CSV file
# We'll name the file 'israel_gdp.csv'
# index=False prevents pandas from writing the DataFrame index as a column in the CSV
israel_gdp.to_csv('/tmp/israel_gdp.csv', index=False)

print("DataFrame saved successfully to 'israel_gdp.csv' in your current working directory.")
```

```python
# Create a line chart for Israel's GDP over the years
plt.figure(figsize=(12, 6))
plt.plot(israel_gdp['Year'], israel_gdp['Value'], marker='1', color='skyblue')

plt.title('Israel GDP Over Years')
plt.xlabel('Year')
plt.ylabel('GDP (USD)')
plt.grid(True)
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
